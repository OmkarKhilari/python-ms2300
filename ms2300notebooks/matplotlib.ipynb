```python
## curve fit
from pandas import read_csv
from scipy.optimize import curve_fit
```

```python
# define the objective function
def objective(x, a, b):
    return a * x + b
```

```python
# load the dataset
url = 'https://raw.githubusercontent.com/jbrownlee/Datasets/master/longley.csv'
dataframe = read_csv(url, header=None)
data = dataframe.values
#data
```

```python
# choose the input and output variables
x, y = data[:, 4], data[:, -1]

# curve fit
popt, pcov = curve_fit(objective, x, y)
# summarize the parameter values
a, b = popt
print('y = %.5f * x + %.5f' % (a, b))
```

```python
# plot input vs output
fig, ax = plt.subplots(figsize=(4.,4.))
ax.scatter(x, y)

# define a sequence of inputs between the smallest and largest known inputs
x_line = np.arange(min(x), max(x), 1)

# calculate the output for the range
y_line = objective(x_line, a, b)

ax.set_xlabel('X', fontsize=18, labelpad=5)
ax.set_ylabel('Y', fontsize=18, labelpad=10)

# create a line plot for the mapping function
ax.plot(x_line, y_line, '--', color='red')
plt.show()
```

```python
#dependent-variable from the fit
yfit = 0.48488 * x + 8.38067
```

```python
yfit
```

```python
## calculating errors from definition

sqerr = 0.0
sumsqerr = 0.0
var = 0.0
sumvar = 0.0
abserr = 0.0
sumabserr = 0.0
mean = np.sum(y)/len(y)

for i in range(len(x)):
    sqerr = np.square(y[i]-yfit[i])
    sumsqerr = sumsqerr + sqerr
    var = np.square(y[i]-mean)
    sumvar = sumvar + var
    abserr = np.abs(y[i]-yfit[i])
    sumabserr = sumabserr + abserr
    
mse = sumsqerr/len(x)
rmse = np.sqrt(mse)
mae = sumabserr/len(x)
sse = sumsqerr #sum of squares residual error
sst = sumvar #sum of squares total = sse + ssr (sum of squares regression)
r2 = 1. - sse/sst 

print('\n mse=', mse,'\n','rmse=', rmse,'\n','mae=', mae, '\n','r2=', r2,)
```

```python
## calculating mse using numpy (short-version)
mse_alt1 = np.square(np.subtract(y,yfit)).mean() 
mse_alt1
```

```python
## Calculating the MSE with sklearn
from sklearn.metrics import mean_squared_error
mse_alt2 = mean_squared_error(y,yfit)
mse_alt2
```

```python
from sklearn.metrics import r2_score
r2_alt = r2_score(y, yfit)
r2_alt
```
